```python
df_Filtrado.info()
```

```python
#filtering the data with >=3 ratings 
df_Filtrado = df_Filtrado[df_Filtrado['ao_hechos'] >= 2016]
```

Muestra de indicadores delictivos por alcaldia de la CDMX

```python
ConteoDelito = (df_Filtrado['delito'].isin([0,1]).groupby(df_Filtrado['alcaldia_hechos']).count()
                     .astype(int).reset_index(name='CuentaDelito'))
ConteoDelito
```

```python
DelitoTemporada = (df_Filtrado['delito'].isin([0,1]).groupby(df_Filtrado['mes_hechos']).count()
                     .astype(int).reset_index(name='MesDelito'))
DelitoTemporada
```

```python
dateTimeObj = datetime.now()
```

```python
timestampStr = dateTimeObj.strftime("%d")
print('Current Timestamp : ', timestampStr)
```

```python
df_Filtrado['fecha_hechos2'] = df_Filtrado['fecha_hechos'].astype(str)
df_Filtrado['DOW'] = df_Filtrado['fecha_hechos2'].str.slice(start=8, stop=10)
df_Filtrado.drop(columns=['fecha_hechos2'])
df_Filtrado['DOW']
```

```python
DelitoDia = (df_Filtrado['delito'].isin([0,1]).groupby(df_Filtrado['DOW']).count()
                     .astype(int).reset_index(name='DiaDelito2'))
DelitoDia
```

```python
df_Filtrado.categoria_delito.unique()
```

```python
DelitoTipo = (df_Filtrado['delito'].isin([0,1]).groupby(df_Filtrado['categoria_delito']).count()
                   .astype(int).reset_index(name='CatDelito'))
DelitoTipo.sort_values(by=['CatDelito'])
```

```python
DelitoAno = (df_Filtrado['delito'].isin([0,1]).groupby(df_Filtrado['ao_hechos']).count()
                   .astype(int).reset_index(name='AnoDelito'))
DelitoAno
```

```python
filAlcaldia = df_Filtrado.groupby(['alcaldia_hechos', 'categoria_delito']).size()
filAlcaldia
```
